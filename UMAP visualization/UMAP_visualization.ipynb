```python
import numpy as np
import umap 
# performances visualization 
import matplotlib.pyplot as plt
```

```python
dataX_A = np.load('dataX_A.npy')
```

```python
new_dataX_A = []
for data in dataX_A:
    new_dataX_A.append(data[0])
```

```python
dataX_C = np.load('dataX_C.npy')
```

```python
new_dataX_C = []
for data in dataX_C:
    new_dataX_C.append(data[0])
```

```python
dataY = np.load('dataY.npy')
```

```python
dataY[0]
```

```python
A_umap = umap.UMAP(n_components=2).fit_transform(new_dataX_A)
plt.figure(figsize=(10,10))
plt.scatter(A_umap[:,0],A_umap[:,1], c = dataY)
plt.show()
```

```python
C_umap = umap.UMAP(n_components=2).fit_transform(new_dataX_C)
plt.figure(figsize=(10,10))
plt.scatter(C_umap[:,0],C_umap[:,1], c = dataY)
plt.show()
```
